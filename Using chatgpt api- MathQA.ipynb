# Table 1- gpt-3.5-turbo

```python
import pandas as pd

table1Question=[]
table1Correct=[]
table1Answer=[]

for i, row in df.iloc[:1000].iterrows():
    table1Question.append("Answer only with the correct letter: " + row["Problem"]+ " "+ row["options"])
    table1Correct.append(row["correct"])
    table1Answer.append("0")
#We have first 500 enteries in table1
```

```python
list_of_tuples = list(zip(table1Question,table1Correct,table1Answer))
pd.set_option('display.max_colwidth', None)
table1 = pd.DataFrame(list_of_tuples, columns=["Question", "Correct", "Answer"])
table1
```

```python
import openai  # for OpenAI API calls
openai.api_key = "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"

from tenacity import (
    retry,
    stop_after_attempt,
    wait_random_exponential,
)  # for exponential backoff


@retry(wait=wait_random_exponential(min=1, max=60), stop=stop_after_attempt(6))
def completion_with_backoff(**kwargs):
    return openai.ChatCompletion.create(**kwargs)
```

```python
for i in range(0, len(table1)):
    ans= completion_with_backoff(
    model = "gpt-3.5-turbo", 
    temperature = 0.8, 
    max_tokens = 2000,
    messages = [{"role": "system", "content": table1.loc[i,"Question"]}]
    )    
    table1.loc[i, ['Answer']] = ans.choices[0].message.content
```

```python
table1
```

```python
for i in range(0, len(table1)):
    x=table1.loc[i, ['Answer']].astype("string")[0]
    print(x)
#We can see there are multiple answers that are not answered using mcq we need to clean the dataset
```

```python
for i in range(0, len(table1)):
    s=table1.loc[i, ['Answer']].astype("string")[0]
    if (s.find('a)')!=-1):
        table1.loc[i, ['Answer']] = 'a'
    elif (s.find('b)')!=-1):
        table1.loc[i, ['Answer']] = 'b'
    elif (s.find('c)')!=-1):
        table1.loc[i, ['Answer']] = 'c'
    elif (s.find('d)')!=-1):
        table1.loc[i, ['Answer']] = 'd'
    elif (s.find('e)')!=-1):
        table1.loc[i, ['Answer']] = 'e'
    else:
        table1.loc[i, ['Answer']] = '0'
table1
```

```python
#table1 has 500 questions solved using gpt-3.5-turbo
#Correct contains the right option and answer contains the option given by the model
table1.to_csv('Math1000GPT3.csv')
```

```python
j=0
for i in range(0, len(table1)):
    ans=(table1.loc[i, ['Answer']][0])
    corr=(table1.loc[i, ['Correct']]).astype("string")[0]
    if(ans==corr):
        j=j+1
print(j)
print((j/1000)*100)
```
